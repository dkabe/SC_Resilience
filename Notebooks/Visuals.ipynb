```python
import math
import numpy as np
import pandas as pd
from random import seed
import matplotlib.pyplot as plt
```

```python
path = "/home/dkabe/Model_brainstorming/"
```

```python
## Instance 1
```

```python
df = pd.read_csv(path + 'Output/results.txt', delimiter = "\t", header=None)
```

```python
df = df[26:36]
```

```python
df
```

```python
rl = [0.5 , 0.55, 0.6 , 0.65, 0.7, 0.75, 0.8 , 0.85, 0.9 , 0.95]
Obj = df[0]
InHouse = df[1]
Outsource = df[2]
LostSales = df[3]
RL_Penalties = df[4]
```

```python
## Costs against different Recovery Levels 
```

```python
plt.plot(rl, Obj)
plt.title("Recovery Level vs. Total Cost")
plt.xlabel("Recovery Level")
plt.ylabel("Total Cost")
plt.show()
```

```python
plt.plot(rl, InHouse)
plt.title("Recovery Level vs. In House Cost")
plt.xlabel("Recovery Level")
plt.ylabel("In House Cost")
plt.show()
```

```python
plt.plot(rl, Outsource)
plt.title("Recovery Level vs. Cost of Outsourcing")
plt.xlabel("Recovery Level")
plt.ylabel("Cost of Outsourcing")
plt.show()
```

```python
plt.plot(rl, LostSales)
plt.title("Recovery Level vs. Cost of Lost Sales")
plt.xlabel("Recovery Level")
plt.ylabel("Cost of Lost Sales")
plt.show()
```

```python
plt.plot(rl, RL_Penalties)
plt.title("Recovery Level vs. Penalties")
plt.xlabel("Recovery Level")
plt.ylabel("Penalties")
plt.show()
```

```python
## Costs against each other
```

```python
plt.scatter(Outsource, LostSales)
plt.title("Cost of Outsourcing vs. Cost of Lost Sales")
plt.xlabel("Cost of Outsourcing")
plt.ylabel("Cost of Lost Sales")
plt.show()
```

```python
plt.scatter(Outsource, RL_Penalties)
plt.title("Cost of Outsourcing vs. Penalties")
plt.xlabel("Cost of Outsourcing")
plt.ylabel("Cost of Penalties")
plt.show()
```

```python
plt.scatter(LostSales, RL_Penalties)
plt.title("Cost of Lost Sales vs. Penalties")
plt.xlabel("Cost of Lost Sales")
plt.ylabel("Cost of Penalties")
plt.show()
```
