```python
import math
import numpy as np
import pandas as pd
from random import seed
import matplotlib.pyplot as plt
```

```python
path = "/home/dkabe/Model_brainstorming/"
```

```python
## Instance 1
```

```python
df = pd.read_csv(path + 'Output/results.txt', delimiter = "\t", header=None)
```

```python
df = df[:10]
df = df.sort_values(by=[0])
```

```python
df
```

```python
rl = [0.5 , 0.55, 0.6 , 0.65, 0.7, 0.75, 0.8 , 0.85, 0.9 , 0.95]
```

```python
Obj = np.array(df[1])/np.array(df[1])
InHouse = np.array(df[2])/np.array(df[1])
Outsource = np.array(df[3])/np.array(df[1])
LostSales = (np.array(df[4]) + np.array(df[5]))/np.array(df[1])
RL_Penalties = np.array(df[5])/np.array(df[1])
```

```python
## Costs against different Recovery Levels 
```

```python
plt.plot(rl, Obj)
plt.title("Recovery Level vs. Total Cost")
plt.xlabel("Recovery Level")
plt.ylabel("Total Cost")
plt.show()
```

```python
plt.scatter(rl, InHouse)
plt.title("Recovery Level vs. In House Cost")
plt.xlabel("Recovery Level")
plt.ylabel("In House Cost")
plt.show()
```

```python
plt.scatter(rl, Outsource)
plt.title("Recovery Level vs. Cost of Outsourcing")
plt.xlabel("Recovery Level")
plt.ylabel("Cost of Outsourcing")
plt.show()
```

```python
plt.scatter(rl, LostSales)
plt.title("Recovery Level vs. Cost of Lost Sales")
plt.xlabel("Recovery Level")
plt.ylabel("Cost of Lost Sales")
plt.show()
```

```python
plt.scatter(rl, RL_Penalties)
plt.title("Recovery Level vs. Penalties")
plt.xlabel("Recovery Level")
plt.ylabel("Penalties")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
fig.subplots_adjust(hspace=0.1)  # adjust space between axes

#ax1 = fig.add_subplot(111)
#ax2 = fig.add_subplot(111)

ax1.plot(rl, InHouse, 'b--', label='In-house cost')
ax1.plot(rl, Outsource, 'r-x', label='Outsource cost')
ax1.plot(rl, LostSales, 'g->', label='Lost Sales and Penalties')

ax2.plot(rl, InHouse, 'b--', label='In-house cost')
ax2.plot(rl, Outsource, 'r-x', label='Outsource cost')
ax2.plot(rl, LostSales, 'g->', label='Lost sales')

ax1.set_ylim(0.7, 1)  # outliers only
ax2.set_ylim(0, 0.25)  # most of the data

ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.axes.xaxis.set_visible(False)
ax1.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.xaxis.tick_bottom()


d = .5  # proportion of vertical to horizontal extent of the slanted line
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)


ax1.legend(loc='upper right')
plt.xlabel('Recovery Level', fontsize= 12)
#plt.ylabel('Objectives', multialignment = 'center')
fig.text(0.03, 0.5, 'Objectives', ha='center', va='center', rotation='vertical', fontsize= 12)

#plt.yticks(np.arange(0, 0.9, 0.05))
#plt.xticks(np.arange(0.5, 1, 0.05))
plt.savefig("/home/dkabe/Model_brainstorming/Images/Instance_1/objectives.eps")
plt.show()
```
