```python
rl = [0.5, 0.75, 0.95]
x_list = []
y_list = []
```

```python
for r_level in rl:
    data = np.loadtxt('/home/dkabe/Model_brainstorming/Epsilon_Constraint/Two_Objectives/Instance_1/e_cons_results_' + str(r_level) + '.txt')
    x = data[:,0]
    y = data[:,1]
    plt.plot(x, y, 'ro')
    plt.xlabel("Cost of met demand")
    plt.ylabel("Cost of unmet demand")
    plt.show()
    x_list.append(x)
    y_list.append(y)
```

```python
plt.scatter(x_list[0], y_list[0], label = 'rl at 0.5')
plt.scatter(x_list[1], y_list[1], label = 'rl at 0.75')
plt.scatter(x_list[2], y_list[2], label = 'rl at 0.95')
plt.xlabel("Cost of met demand")
plt.ylabel("Cost of unmet demand")
plt.legend()
plt.show()
```
